```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the cv2_imshow function from google.colab.patches

# Read the input image
image = cv2.imread('/content/sample_data/princess.jpg')

# 1. Resize the image (optional)
resized_image = cv2.resize(image, (300, 300))  # Resize to 300x300 pixels

# 2. Convert the image to grayscale
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# 3. Apply Gaussian Blur to reduce noise
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# 4. Apply Thresholding (Otsu's method for automatic thresholding)
_, threshold_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 5. Canny Edge Detection
edges = cv2.Canny(threshold_image, 100, 200)

# 6. Dilation to expand edges
dilated_image = cv2.dilate(edges, (3, 3), iterations=1)

# 7. Erosion to shrink edges (optional)
eroded_image = cv2.erode(dilated_image, (3, 3), iterations=1)

# Display the images using cv2_imshow instead of cv2.imshow
cv2_imshow(image) # Use cv2_imshow to display the images in Colab
cv2_imshow(resized_image)
cv2_imshow(gray_image)
cv2_imshow(blurred_image)
cv2_imshow(threshold_image)
cv2_imshow(edges)
cv2_imshow(dilated_image)
cv2_imshow(eroded_image)

```
